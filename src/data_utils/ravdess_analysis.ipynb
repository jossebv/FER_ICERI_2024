```python
import pandas as pd
import os
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt

DATA_PATH = "../../data/RAVDESS/videos/"

videos_lenghts = []
max_length = 0
for actor in tqdm(os.listdir(DATA_PATH), desc="Processing actors"):
    actor_path = os.path.join(DATA_PATH, actor)
    for video in os.listdir(actor_path):
        cap = cv2.VideoCapture(os.path.join(actor_path, video))
        n_frames = 0
        while True:
            ret, frame = cap.read()
            if not ret:
                break

            n_frames += 1

        if n_frames > max_length:
            max_length = n_frames
        videos_lenghts.append(n_frames/30)
        

plt.hist(videos_lenghts, bins=30)
print(f"Max lenght: {max_length}")
```
